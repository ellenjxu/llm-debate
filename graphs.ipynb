```python
# fig 1: math accuracy vs agent number
import os

def get_results_from_txt(result_dir, fig="agents"):
    results = []

    for result_file in os.listdir(result_dir):
        if result_file.endswith(".txt"):
            with open(os.path.join(result_dir, result_file), "r") as f:
                lines = f.readlines()
                last_line = lines[-1]
                accuracy, std = last_line.split(",")
                if fig == "agents": 
                    agent_number = os.path.basename(result_file).split("_")[1][-1]
                elif fig == "rounds":
                    agent_number = os.path.basename(result_file).split("_")[2].split(".")[0][-1]
                elif fig == "panel":
                    agent_number = os.path.basename(result_file).split(".")[0][-5:]
                    # calculate diversity, ex how many different models there are in agent_number
                    agent_number = len(set(agent_number.split("_")))
                results.append((agent_number, float(accuracy)*100, float(std)*100))
    return results
```

```python
# make a plot of accuracy vs agent number
def generate_graph(results, fig="agents"):
    results_int = [(int(agent_number), float(accuracy), float(std)) for agent_number, accuracy, std in results]
    results_sorted = sorted(results_int, key=lambda x: x[0])

    # Unpack the sorted results
    agent_numbers, accuracies, stds = zip(*results_sorted)

    plt.figure(figsize=(5, 4))
    # plt.errorbar(agent_numbers, accuracies, yerr=stds, fmt='-o', capsize=5, capthick=2, ecolor='red', marker='s', markersize=5, linestyle='--', linewidth=2, label='Accuracy vs Agent Number')
    plt.plot(agent_numbers, accuracies, '-o', markersize=5, linestyle='--', linewidth=1, label='Accuracy vs Agent Number', color='black')
    if fig == "agents":
        plt.title('Math Accuracy vs Number of Debating Agents')
        plt.xlabel('Agent Number')
    elif fig == "rounds":
        plt.title('Math Accuracy vs Debate Rounds')
        plt.xlabel('Debate Rounds')
    elif fig == "panel":
        plt.title('Math Accuracy vs Diversity of Debating Agents')
        plt.xlabel('Diversity of Debating Agents')
    plt.ylabel('Math Accuracy')
    plt.xticks(agent_numbers)
    # plt.legend()
    plt.grid(True)
    plt.show()
```

```python
result_dir = "output/agents_results"
agents_results = get_results_from_txt(result_dir)
generate_graph(agents_results)
```

```python
rounds_dir = "output/rounds_results2"
rounds_results = get_results_from_txt(rounds_dir, "rounds")
generate_graph(rounds_results, "rounds")
```

```python
panel_dir = "output/panel_results"
panel_results = get_results_from_txt(panel_dir, "panel")
generate_graph(panel_results, "panel")
```

```python
# bar plot

results = {}
results_dir = "output/panel_results"
target_models = ['0_0_0', '1_1_1', '2_2_2', '0_1_2']
display_names = {'0_0_0': 'Mistral', '1_1_1': 'Llama', '2_2_2': 'Gemma', '0_1_2': 'Panel'}
    
for result_file in os.listdir(results_dir):
    if result_file.endswith(".txt"):
        model_id = os.path.basename(result_file).split(".")[0][-5:]
        if model_id in target_models:
            with open(os.path.join(results_dir, result_file), "r") as f:
                lines = f.readlines()
                last_line = lines[-1]
                accuracy, _ = last_line.split(",")
                display_name = display_names[model_id]
                results[display_name] = float(accuracy)*100

# sorted_results = sorted(results.items(), key=lambda x: x[0])
model_ids = results.keys()
accuracies = results.values()
    
plt.figure(figsize=(5, 4))
bars = plt.bar(model_ids, accuracies, color='grey', linewidth=1, width=0.5)

colors = ['0.5', '0.8', '0.8', '0.8']
for bar, shade in zip(bars, colors):
    bar.set_facecolor(shade)

plt.xlabel('Debating Agent Models')
plt.ylabel('Math Accuracy')
plt.title('Math Accuracy vs Diversity of Debating Agents')
plt.xticks(rotation=45)  # Rotate model IDs for better readability

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/4, yval, round(yval, 2), va='bottom')  # va: vertical alignment

plt.tight_layout()
plt.show()
```
